```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
```

```python
movie_df = pd.read_csv("movie.csv")
rating_df = pd.read_csv("rating.csv")
```

```python
df = rating_df.merge(movie_df,how="left",on="movieId")
```

```python
df.head()
```

```python
a = (df.groupby("movieId").agg({"rating": "count"}).rating<=10000)
```

```python
movieList = list(a[a==True].index)
```

```python
len(movieList)
```

```python
df.shape
```

```python
df = df[~df["movieId"].isin(movieList)]
```

```python
df.shape
```

```python
user_movie_df = df.pivot_table(index=["userId"],columns=["title"],values = "rating")
```

```python
user_movie_df.head()
```

```python
def createUserMovieDf(rating_df,movie_df,co_count=10000):
    df = rating_df.merge(movie_df, how="left", on="movieId")
    comment_counts = df["title"].value_counts()
    rare_movies = comment_counts[comment_counts<=co_count].index
    common_movies = df[~df["title"].isin(rare_movies)]
    user_movie_df = common_movies.pivot_table(index=["userId"],columns=["title"],values = "rating")
    return user_movie_df
```

```python
random_user_df = user_movie_df.sample(1)
```

```python
userId = 47706
```

```python
movies_watched = random_user_df.columns[random_user_df.notna().any()].tolist()
```

```python
movies_watched_df = user_movie_df[movies_watched]
```

```python
movies_watched_df.head()
```

```python
user_movie_count = movies_watched_df.T.notnull().sum()
```

```python
user_movie_count = user_movie_count.reset_index()
```

```python
user_movie_count.columns = ["userId", "movie_count"]
```

```python
user_movie_count.head()
```

```python
perc = len(movies_watched)*60/100
```

```python
users_same_movies = user_movie_count[user_movie_count["movie_count"]> perc]["userId"]
```

```python
len(users_same_movies)
```

```python
final_df = movies_watched_df[movies_watched_df.index.isin(users_same_movies)]
```

```python
final_df.head()
```

```python
corr_df = final_df.T.corr().unstack().sort_values()
```

```python
corr_df = pd.DataFrame(corr_df, columns=["corr"])
```

```python
corr_df.index.names = ["user_id_1", "user_id_2"]
```

```python
corr_df = corr_df.reset_index()
```

```python
corr_df[corr_df["user_id_1"]==userId]
```

```python
top_users = corr_df[(corr_df["user_id_1"]== userId)&(corr_df["corr"]>=0.65)][["user_id_2","corr"]].reset_index(drop=True)
```

```python
top_users = top_users.sort_values("corr", ascending=False)
```

```python
top_users.rename(columns={"user_id_2": "userId"}, inplace=True)
```

```python
top_users_ratings = top_users.merge(rating_df[["userId","movieId","rating"]],how="inner")
```

```python
top_users_ratings = top_users_ratings[top_users_ratings["userId"] != userId]
```
