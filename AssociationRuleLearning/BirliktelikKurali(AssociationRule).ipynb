# Birliktelik Kuralı Öğrenimi(Association Rule Learning)

Veri içerisindeki örüntüleri, ilişkileri (pattern, ilişki, yapı) bulmak için
kullanılan kural tabanlı bir *makine Öğrenmesi* tekniğidir.

### Örnek
Mehmet bey bir alışveriş yapmış olsun onun bir alışveriş sepeti ve fişi var
Ayşe hanım farklı bir alışveriş yapmış olsu onunda sepeti ve fişi var
ve böyle toplam da örneğin 8 kişi alışveriş yapmış.
Bu müşterilerin alışveriş yaptıktan sonra fişlerini alsak ve incelesek
örneğin birlikte gözlemlenen ürün çiftlerini not etsek.
Örneğin fişlere baktık ve şunu gördük 8 tane müşteriden 5 tanesi
ekmek ve sütü birlikte almış. Bu 8 alışveriş fişi göz önünde bulundurulduğunda ekmekle sütün birlikte gözlenme oranı 5/8 dir.
Yani şunu diyebiliriz : Alışverişlerin yaklaşık olarak %62 sinde ekmek ve süt birlikte gözlemlenmiştir. 
* Bu örnek bir birliktelik kuralı örneğidir

## Apriori Algoritması

* Sepet Analizi yöntemidir.
* Ürün birlikteliklerini ortaya çıkarmak için kullanılır.
* 3 temel metrik vardır bu 3 temel metrik ile veri seti içerisindeki ilişki örüntülerini gözlemleme imkanı buluruz.
* Bu ilişkileri istatistiksel bir ölçüt ile değerlendirme imkanı buluruz.



* ################## Support(X,Y) = Freq(X,Y)/N ###################
* ###############  X ve Y nin birlikte görülme olasılığı  ##############

* ################## Confidence(X,Y) = Freq(X,Y)/Freq(X) ###################
* ##############  X satın alındığında Y'nin satılması olasılığı ##############


* ################## Lift = Support(X,Y)/ (Support(X) * Support(Y))###################
* ############  X satın alındığında Y'nin satın alınma olasılığı lift kadar artar  ###########
