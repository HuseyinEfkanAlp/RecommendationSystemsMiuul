# İçerik Temelli Filtreleme (Content Based Filtering)


**Ürün içeriklerinin benzerlikleri üzerinden tavsiyeler geliştirir.**

Elimizdeki ürünün meta bilgileri üzerinden benzerlikler hesaplanır ve o
ilgili ürüne **en benzer olan ürünler önerilir**

## Benzerliği, Yakınlığı nasıl ölçebilirim?

1. Metinleri Matematiksel Olarak Temsil Et (Metinleri Vektörleştirme)
2. Benzerlikleri Hesapla

### Metinleri Vektörleştirmek

Bu yöntemleri kullanarak elimizdeki metinleri vektörleştirebiliriz
* Count Vector (Word Count)
* TF-IDF

### Benzerlikleri Hesaplama

1. Euclidean Distance (Öklid Uzaklığı):

	İki vektörün birbirleri arasındaki ***uzaklık***ların karelerinin toplamlarının kare kökü.

2. **Cosine Similarity** (Kosinüs Benzerliği)

	İki vektörün birbirleri arasındaki ***benzerlik***leri

Euclidean uzaklık Cosine benzerlik hesaplar

## Metin Vektörleştirmek
###  1) Count Vector

1. Adım: Eşsiz Tüm Terimleri Sütunlara, Bütün Dökümanları Satırlara Yerleştir

2. Adım: Terimlerin Dökümanlarda Geçme Frekanslarını Hücrelere Yerleştir

### 2) TF-IDF

1. Adım: Count Vectorizer'ı Hesapla

2. Adım. TF- Term Frequency'yi Hesapla

(t teriminin ilgili dökümandaki frekansı / dökümandaki toplam terim sayısı)

3. Adım: IDF - Inverse Document Frequency'i Hesapla

1+loge((toplam döküman sayısı + 1)/(içinde t terimi olan döküman sayısı + 1))

4. Adım: TF*IDF'i Hesapla

5. Adım: L2 Normalizasyonu Yap

Satırların kareleri toplamının karekökünü bul, ilgili satırdaki tüm hücreleri bulduğun değere böl
